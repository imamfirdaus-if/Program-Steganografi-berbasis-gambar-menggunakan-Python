# **Penjelasan Steganografi**
Steganografi adalah metode menyembunyikan data rahasia dalam gambar/audio/video. Singkatnya, motif utama steganografi adalah untuk menyembunyikan informasi yang dimaksud dalam gambar/audio/video yang tidak tampak rahasia hanya dengan melihatnya.
Ide di balik Steganografi berbasis gambar sangat sederhana. Gambar terdiri dari data digital (piksel), yang menjelaskan apa yang ada di dalam gambar, biasanya warna semua piksel. Karena kita tahu setiap gambar terdiri dari piksel dan setiap piksel mengandung 3 nilai (merah, hijau, biru).

# **Proses Encode data**
Setiap byte data diubah menjadi kode biner 8-bit menggunakan nilai ASCII. Sekarang piksel dibaca dari kiri ke kanan dalam kelompok 3 yang berisi total 9 nilai. 8-nilai pertama digunakan untuk menyimpan data biner. Nilai dibuat ganjil jika 1 terjadi dan genap jika 0 terjadi.
Contoh :
Misalkan pesan yang akan disembunyikan adalah 'Hii'. Karena pesan berukuran 3 byte, oleh karena itu, piksel yang diperlukan untuk mengkodekan data adalah 3 x 3 = 9. Pertimbangkan gambar 4 x 3 dengan total 12 piksel, yang cukup untuk mengkodekan data yang diberikan.

# **Proses Dekode data**
Untuk memecahkan kode, tiga piksel dibaca sekaligus, hingga nilai terakhir ganjil, yang berarti pesan telah selesai. Setiap 3-piksel berisi data biner, yang dapat diekstraksi dengan logika pengkodean yang sama. Jika nilainya jika ganjil bit binernya adalah 1 else 0.

```python
# Program Python mengimplementasikan Steganografi Gambar

# Modul PIL digunakan untuk mengekstrak piksel gambar dan memodifikasinya
from PIL import Image

# Ubah data penyandian menjadi biner 8-bit
# bentuk menggunakan nilai karakter ASCII
def genData(data):

		# daftar kode biner dari data yang diberikan
		newd = []

		for i in data:
			newd.append(format(ord(i), '08b'))
		return newd

# Piksel dimodifikasi sesuai dengan data biner 8-bit dan akhirnya dikembalikan
def modPix(pix, data):

	datalist = genData(data)
	lendata = len(datalist)
	imdata = iter(pix)

	for i in range(lendata):

		# Mengekstrak 3 piksel sekaligus
		pix = [value for value in imdata.__next__()[:3] +
								imdata.__next__()[:3] +
								imdata.__next__()[:3]]

		# Nilai piksel harus dibuat ganjil untuk 1 dan genap untuk 0
		for j in range(0, 8):
			if (datalist[i][j] == '0' and pix[j]% 2 != 0):
				pix[j] -= 1

			elif (datalist[i][j] == '1' and pix[j] % 2 == 0):
				if(pix[j] != 0):
					pix[j] -= 1
				else:
					pix[j] += 1
				# pix[j] -= 1

		# Piksel kedelapan dari setiap set memberi tahu apakah akan berhenti atau membaca lebih lanjut.
		#0 berarti terus membaca
		#1 berarti pesan sudah selesai.
		if (i == lendata - 1):
			if (pix[-1] % 2 == 0):
				if(pix[-1] != 0):
					pix[-1] -= 1
				else:
					pix[-1] += 1

		else:
			if (pix[-1] % 2 != 0):
				pix[-1] -= 1

		pix = tuple(pix)
		yield pix[0:3]
		yield pix[3:6]
		yield pix[6:9]

def encode_enc(newimg, data):
	w = newimg.size[0]
	(x, y) = (0, 0)

	for pixel in modPix(newimg.getdata(), data):

		# Menempatkan piksel yang dimodifikasi di gambar baru
		newimg.putpixel((x, y), pixel)
		if (x == w - 1):
			x = 0
			y += 1
		else:
			x += 1

# Encode data menjadi gambar
def encode():
	img = input("Masukkan nama file beserta extensionnya(contoh: image.png) : ")
	image = Image.open(img, 'r')

	data = input("Masukkan kalimat yang akan di encode : ")
	if (len(data) == 0):
		raise ValueError('Kalimat belum dimasukkan')

	newimg = image.copy()
	encode_enc(newimg, data)

	new_img_name = input("Masukkan nama file baru beserta extensionnya(contoh: newimage.png) : ")
	newimg.save(new_img_name, str(new_img_name.split(".")[1].upper()))

# Decode data dalam gambar
def decode():
	img = input("Masukkan nama file beserta extensionnya(contoh: image.png) : ")
	image = Image.open(img, 'r')

	data = ''
	imgdata = iter(image.getdata())

	while (True):
		pixels = [value for value in imgdata.__next__()[:3] +
								imgdata.__next__()[:3] +
								imgdata.__next__()[:3]]

		# string data biner
		binstr = ''

		for i in pixels[:8]:
			if (i % 2 == 0):
				binstr += '0'
			else:
				binstr += '1'

		data += chr(int(binstr, 2))
		if (pixels[-1] % 2 != 0):
			return data

# Fungsi Utama
def main():
	a = int(input(":: Program Steganografi berbasis gambar menggunakan Python::\n"
						"Pilih menu (1 atau 2)\n1. Encode\n2. Decode\n"))
	if (a == 1):
		encode()

	elif (a == 2):
		print("Kata yang Didekode : " + decode())
	else:
		raise Exception("Masukkan kalimat yang benar")

if __name__ == '__main__' :

	# Memanggil fungsi utama
	main()
```
